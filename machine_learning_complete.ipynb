```python
house = pd.read_csv("C:/Users/suraj/data_visualisation/Housing.csv")
df3 = pd.DataFrame(house)
```

```python
from sklearn.preprocessing import StandardScaler

from sklearn.model_selection import train_test_split
```

```python
house.head(10)
```

```python
sns.heatmap(data=df3.select_dtypes(include='number').corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

```python
x = house.iloc[:,:-1]
y =house["price"]
```

```python
from sklearn.preprocessing import StandardScaler

# Keep only numeric columns
x_numeric = x.select_dtypes(include='number')

sc = StandardScaler()
sc.fit_transform(x_numeric)
```

```python

x= pd.DataFrame(sc.transform(x_numeric), columns=x_numeric.columns)
x
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, train_size=0.2 , random_state=42)
```

```python
from sklearn.linear_model import LinearRegression, Lasso,Ridge
import numpy as np
from sklearn.metrics import mean_absolute_error, mean_squared_error
```

Linear Regression

```python
lr3 = LinearRegression()
lr3.fit(x_train, y_train)
lr3.score(x_test, y_test)*100
```

```python
print(mean_squared_error(y_test,lr3.predict(x_test)))
print(mean_absolute_error(y_test, lr3.predict(x_test)))
print(np.sqrt(mean_squared_error(y_test,lr3.predict(x_test))))
```

```python
lr3.coef_
```

```python
lr3.intercept_
```

```python
plt.bar(x_numeric.columns, lr3.coef_)
plt.xlabel("colmns")
plt.ylabel("coef")
```

Lasso

```python
la= Lasso(alpha=0.0001)
la.fit(x_train,y_train)
la.score(x_test, y_test)
```

```python
print(mean_squared_error(y_test,la.predict(x_test)))
print(mean_absolute_error(y_test, la.predict(x_test)))
print(np.sqrt(mean_squared_error(y_test,la.predict(x_test))))
```

```python
plt.bar(x_numeric.columns, la.coef_)
plt.xlabel("colmns")
plt.ylabel("coef")
```

Ridge

```python
ra = Ridge(alpha=0.001)
ra.fit(x_train,y_train)
ra.score(x_test, y_test)
```

```python
print(mean_squared_error(y_test,ra.predict(x_test)))
print(mean_absolute_error(y_test, ra.predict(x_test)))
print(np.sqrt(mean_squared_error(y_test,ra.predict(x_test))))
```

```python
plt.bar(x_numeric.columns, ra.coef_)
plt.xlabel("colmns")
plt.ylabel("coef")
```

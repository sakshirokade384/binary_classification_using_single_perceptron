```python
# bar plot
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
data  = sns.load_dataset("tips")
print(data)
sns.barplot(data=data, x ="day", y="tip" , estimator ="mean", hue ="sex", palette="viridis", errorbar=("ci",0))
plt.show()
```

```python
#histogram
data = sns.load_dataset("tips")
sns.histplot(data, x ="day",hue = "sex", kde=True)
plt.show()
```

```python
data = sns.load_dataset("planets")
sns.histplot(data =data, x ="year", kde =True)
plt.show()
```

```python
data = sns.load_dataset("titanic")
sns.histplot(data =data, x ="age", kde =True, hue ="who",bins =10)
plt.show()
```

```python
#scatter plot
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
data = sns.load_dataset("tips")
print(data)
sns.scatterplot(data= data, x = "total_bill", y= "tip", hue = "day", size = "size", marker="D")
plt.legend(bbox_to_anchor =(0.8,0.8,1,1))
plt.show()
```

```python
#Heatmap
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
data = sns.load_dataset("tips")
print(data)
gp = data.groupby("sex").agg({"tip":"mean"})
sns.heatmap(gp)
```

```python
# count plot
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

data = sns.load_dataset("tips")
print(data)
df = pd.DataFrame(data)
df1 = df.head(20)
sns.countplot(data= df1, x="tip", palette="viridis")
plt.show()
```

```python
# voilin plot
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
data = sns.load_dataset("tips")
print(data)
sns.violinplot(data = data, x = "day")
```

```python
#pair plot
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
data = sns.load_dataset("tips")
print(data)
sns.pairplot(data = data)
```

```python
#strip plot
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
data = sns.load_dataset("tips")
print(data)
sns.stripplot(data = data, x="total_bill", y= "time", dodge=True, jitter=0.1)
```

```python
# box plot
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
data = sns.load_dataset("iris")
print(data)
import seaborn as sns
sns.boxplot(x="sepal_width", y="sepal_length", data=data, orient="v", fliersize = 10)
plt.show()
```

```python
#categorical plot
data = sns.load_dataset("tips")
print(data)
sns.catplot(x ="tip", y = "day", data = data, kind = "box")
plt.show()
```

```python
#style and color in plot
data = sns.load_dataset("tips")
print(data)
sns.set_style(style="whitegrid")
sns.lineplot(data = data, x= "day", y ="tip")

```

```python
import seaborn as sns
import matplotlib.pyplot as plt

data = sns.load_dataset("tips")
print(data)
sns.palplot(sns.color_palette ("viridis", 4))
plt.show()
```

```python
#multiple plots in seaborn
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
data = sns.load_dataset ("tips")
print(data)
a=  sns.FacetGrid(data, col="smoker",height = 3 )
a.map(sns.barplot,"day", data = data)
plt.show()
```

```python
#relational plot
data = sns.load_dataset ("tips")
print(data)
sns.relplot(data=data, x="day", y="tip", hue = "sex", kind="line")
plt.show()
```

```python
#swarm plot
data = sns.load_dataset ("tips")
print(data)
sns.swarmplot(data= data, x= "sex", y="tip", dodge=True, hue = "day" )
plt.show()

```

```python
#kde plot
data = sns.load_dataset ("tips")
print(data)
sns.kdeplot(data= data, x="tip", multiple="stack", hue= "sex")
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split

# Read the dataset from an Excel file
dataset = pd.read_excel(r"ploynomial_input.csv.xlsx")

# Display the first 3 rows of the dataset
print(dataset.head(3))
```

```python
sns.scatterplot(x="data1",y="data2",data=dataset,hue="output")
plt.show()
```

```python
x=dataset.iloc[:,:-1]
y=dataset["output"]
```

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
pf=PolynomialFeatures(degree=3)
pf.fit(x)
x=pd.DataFrame(pf.transform(x))
```

```python
x, x.shape
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
lr.fit(x_train, y_train)
```

```python
lr.score(x_test,y_test)*100
```

```python
dataset =  pd.read_csv( r"C:\Users\suraj\data_visualisation\placement.csv")
dataset.head(10)
```

```python
x = dataset[["cgpa"]]
y = dataset[["package"]]
```

```python
sns.scatterplot(x = "cgpa",y= "package", data = dataset )
plt.show()
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=2)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr1 = LinearRegression()
```

```python
lr1.fit(x_train,y_train)
# this provides the value of m and c
```

```python
lr1.score(x_test, y_test)*100
```

```python
lr1.predict([[6.89]])
# here predict function takes the value of x from user puts that with values given by fit for m and c and guves the prediction
```

```python
lr1.coef_
# which is m from y=mx+c
```

```python
lr1.intercept_
# which is c from above formula
```

```python
y= (lr1.coef_* 6.89)+lr1.intercept_
y
```

```python
y_predict = lr1.predict(x)
```

```python
sns.scatterplot(data= dataset, x="cgpa", y ="package", label = "original_data")
plt.plot(x, y_predict, c= "red",label = "Prediction_line")
plt.legend()
plt.show()
```

```python
sample_mean = []  # empty list

for _ in range(50):
    sample = np.random.choice(p_data, size=500)
    mean_val = np.mean(sample)
    sample_mean.append(mean_val)  # append each sample mean
```

```python
sns.kdeplot(x="sample_mean", data= sample_m)
plt.show()
```

chi square test


```python
ob = np.array([22,17,20,26,22,13])
ex = np.array([20,20,20,20,20,20])
```

```python
chi_square = np.sum((ob-ex)*(ob-ex)/ex)
print(chi_square)
```

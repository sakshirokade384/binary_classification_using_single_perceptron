```python
groupby = df.groupby(["listed_in(type)", "name"])["approx_cost(for two people)"].sum()
top10 = groupby.sort_values(ascending=False).head(5)

# Convert MultiIndex to string for labels
labels = [' | '.join(tup) for tup in top10.index]

plt.barh(labels, top10.values)
plt.xlabel("Approx Cost (for two people)")
plt.show()
```
